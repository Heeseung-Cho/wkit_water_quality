# 타겟 내용 정리
- 각 타겟값들이 어떤 내용인지 구글링해서 정리하기
- 가능하면 단위도


-  2-MIB (2-Methylisoborneol): 
> - 정의: 수원에 조류가 존재하는지를 나타내는 지표
> - 단위: 리터당 나노그램(ng/L) 또는 리터당 마이크로그램(µg/L) 

- Geosmin: 
> - 정의: 자연 발생 유기 화합물
> - 요인: 토양에 서식하는 박테리아에 의해 생성
> - 효과: 독특한 흙 냄새 
> - 단위:  ng/L(리터당 나노그램) 또는 ppb(10억분의 1)

- Mn (Manganese)
> - 정의: 원자번호 25번의 금속원소인 망간(Manganese)의 화학기호
> - 요인: 자연발생
> - 작용: 유기체의 필수 영양소
> - 단위:  리터당 밀리그램(mg/L) 또는 리터당 마이크로그램(µg/L)

- TOC (Total Organic Carbon)
> - 정의: 물 샘플에 존재하는 유기 탄소의 양
> - 용도: 수질 지표, 유해 오염 물질의 존재 가능성을 측정하는 지표
> - 단위: 일리터당 탄소 밀리그램(mg/L) 또는 리터당 탄소 마이크로그램(μg/C/L)

- blue_green_algae (남조류 또는 청록색조류 또는 시아노박테리아)
> - 정의: 담수 환경에서 꽃을 형성할 수 있는 미세한 광합성 박테리아의 일종 
> - 작용: 인간의 건강과 수중 생물에 위협이 될 수 있는 독소를 생성
> - 작용2: 식수에서 맛과 냄새 문제를 유발
> - 작용3: 수처리 시설을 막을 수 있음
> - 작용4: 지구 탄소 및 질소 순환에서 중요한 역할
> - 단위:  밀리리터당 세포수(cells/mL) 또는 입방미터당 세포수(cells/m^3)


- diatomeae (구조류 또는 diatom)
> - 정의: 담수(淡水)와 해수에 널리 분포하는 플랑크톤
> - 작용1: 대기로부터 이산화탄소를 흡수하고 저장
> - 작용2: 많은 수생 생물의 주요 식품 공급원
> - 작용3: 퇴적암에 잘 보존 (특유의 세포벽 때문)
> - 단위: 물, 바이오매스 또는 엽록소 함량의 단위 부피당 세포 수로 측정 또는 물 샘플에 있는 총 식물성 플랑크톤의 비율 또는 백분율

- synedra (구조류(시네드라))
> - 정의: 담수와 염수에서 발견되는 단세포 유기체 
> - 작용1: 동물성 플랑크톤 및 기타 유기체의 주요 먹이 공급원 역할
> - 작용2: (수질의 중요한 지표)
> - 단위: 밀리리터당 세포수(cells/mL) 또는 세제곱센티미터당 세포수(cells/cm3)
