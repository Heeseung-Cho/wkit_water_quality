### Plotting

```python
## 한번만
data_sample1["measure_date"] = data_sample1["measure_date"].map(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M"))
data_sample2["measure_date"] = data_sample2["measure_date"].map(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M"))
data_sample3["measure_date"] = data_sample3["measure_date"].map(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M"))
```

```python
sns.lineplot(x = "measure_date", y = "turbidity",  data = data_sample1)
sns.lineplot(x = "measure_date", y = "pH",  data = data_sample1)
sns.lineplot(x = "measure_date", y = "TOC",  data = data_sample1)
plt.legend(labels = ["turbidity","pH", "TOC"], loc='center left', bbox_to_anchor=(1,0.5))
plt.ylabel("Value")
```

```python
sns.lineplot(x = "measure_date", y = "turbidity",  data = data_sample2)
sns.lineplot(x = "measure_date", y = "pH",  data = data_sample2)
plt.legend(labels = ["turbidity","pH"], loc='center left', bbox_to_anchor=(1,0.5))
plt.ylabel("Value")
```
